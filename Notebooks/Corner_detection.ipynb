```python
#Harris Corner Detection
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chess = cv2.imread('Data/flat_chessboard.png')
chess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(chess)
```

```python
grey_chess= cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_chess,cmap='gray')
```

```python
real_chess = cv2.imread('Data/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
grey_real_chess= cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_real_chess,cmap='grey')
```

```python
grey_chess = np.float32(grey_chess)
```

```python
dst = cv2.cornerHarris(src=grey_chess,blockSize=2,ksize=3,k=0.04)
```

```python
dst= cv2.dilate(dst,None)
```

```python
chess[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.imshow(chess)
```

```python
grey_real_chess = np.float32(grey_real_chess)
```

```python
dst = cv2.cornerHarris(src=grey_real_chess,blockSize=2,ksize=3,k=0.04)
```

```python
dst= cv2.dilate(dst,None)
```

```python
real_chess[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.imshow(real_chess)
```

```python
#Shi-Tomasi Detection
```

```python
chess = cv2.imread('Data/flat_chessboard.png')
chess = cv2.cvtColor(chess,cv2.COLOR_BGR2RGB)
grey_chess= cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
real_chess = cv2.imread('Data/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
grey_real_chess= cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(grey_real_chess,150,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(real_chess)
```
