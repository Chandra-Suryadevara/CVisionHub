```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img =cv2.imread('Data/puppy.jpg').astype(np.uint8)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img=load_img()
display_img(img)
```

```python
edges = cv2.Canny(image=img,threshold1=100,threshold2=110)
```

```python
plt.imshow(edges)
```

```python
med_value=np.median(img)
```

```python
print(med_value)
```

```python
lower=int(max(0,0.7*med_value))
higher=int(min(255,1.2*med_value))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=higher)
```

```python
plt.imshow(edges)
```

```python
blurred_image=cv2.blur(img,ksize=(4,4))
```

```python
edges = cv2.Canny(image=blurred_image,threshold1=lower,threshold2=higher)
```

```python
plt.imshow(edges)
```
